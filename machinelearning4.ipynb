```python
import pandas as pd
```

```python
x=pd.read_csv('https://raw.githubusercontent.com/codebasics/py/master/ML/14_naive_bayes/titanic.csv')
x.head()
```

```python
dropped=x.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis=1)
dropped
```

```python
target=x.Survived
inputs=dropped.drop(['Survived'],axis=1)
inputs
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
inputs['Sex']=le.fit_transform(inputs['Sex'])
inputs
```

```python
inputs.columns[inputs.isna().any()]
```

```python
inputs.Age=inputs.Age.fillna(inputs.Age.mean())
inputs
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(inputs,target,test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
```

```python
model.fit(X_train,Y_train)
```

```python
model.score(X_test,Y_test)
```
