```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
import numpy as np
```

```python
import pandas as pd
```

```python
x=pd.read_csv("https://raw.githubusercontent.com/H16902/Data/main/Copy%20of%20Global_Superstore2.csv")
x
```

```python
geet=x
y=pd.DatetimeIndex(geet['Order Date']).year.unique()
y
```

```python
furniture=geet[geet["Category"]=="Furniture"]
furniture.describe()
```

```python
tech=geet[geet["Category"]=="Technology"]
tech.describe()
```

```python
office=geet[geet["Category"]=="Office Supplies"]
office.describe()
```

```python
for z in x.columns:
    y=pd.isnull(x[z])
    if len(x[y])!=0:
        print(z,"Column contains missing values")
```

```python
duplicate = geet[geet.duplicated()]
print("there are {} duplicate rows in the file".format(len(duplicate)))
```

```python
x['Order Date']=pd.to_datetime(x['Order Date'])
x['Ship Date']=pd.to_datetime(x['Ship Date'])
```

```python
y=x.groupby("Category")
z=y["Sales"].sum()
plt.bar(z.keys(),z.values)
plt.xlabel("Category")
plt.ylabel("Sales")
plt.title("Sales by Category")
plt.show();
```

```python
x["Region"].unique()
```

```python
pivottable=pd.pivot_table(x,index=pd.DatetimeIndex(geet['Order Date']).year,values="Sales",columns="Category",aggfunc = np.sum)
pivottable
```

```python
pivottable.plot.bar()
plt.xlabel("Year")
plt.title("Sales by Category per year")
plt.show()
```

```python
y=x.groupby("Category")
z=y["Quantity"].sum()
plt.bar(z.keys(),z.values)
plt.xlabel("Category")
plt.ylabel("Quantity")
plt.title("Quantity sold in each Category")
plt.show();
```

```python
pivot_new_table=pd.pivot_table(x,index=pd.DatetimeIndex(geet['Order Date']).year,values="Quantity",columns="Category",aggfunc = np.sum)
pivot_new_table
```

```python
pivot_new_table.plot.bar()
plt.xlabel("Year")
plt.title("Quantity of each Category sold per year")
plt.show()
```

```python
y=x.groupby("Region")
z=y["Profit"].sum()
new = pd.DataFrame.from_dict(z)
new['Profit']
new=new.sort_values('Profit',ascending=False)
new
```

```python
z.keys()
```

```python
plt.bar(new.index,new['Profit'])
plt.xlabel("Region")
plt.ylabel("Profit")
plt.title("Profit in each Region")
plt.xticks(rotation=60)

plt.show();
```

```python
x["Sub-Category"].unique()
```

```python
pivot_new_2_table=pd.pivot_table(x,index=pd.DatetimeIndex(geet['Order Date']).year,values="Profit",columns="Category",aggfunc = np.sum)
pivot_new_2_table
```

```python
pivot_new_3_table=(pivot_new_2_table/pivottable)*100
pivot_new_3_table
```

```python
pivot_new_3_table.plot.bar()
plt.xlabel("Year")
plt.ylabel("Profit Margin(in percentage)")
plt.title("Profit Margin of each Category per year")
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.show();
```

```python
pivot_new_2_table.plot.bar()
plt.xlabel("Year")
plt.ylabel("Profit ")
plt.title("Profit of each Category per year")
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.show();
```
